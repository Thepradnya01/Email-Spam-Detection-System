```python
unemployement= data[["States","Region","Estimated Unemployement Rate"]]
figure=px.sunburst(unemployement,path=["Region","States"],values="Estimated Unemployement Rate",width=700,height=700,color_continuous_scale="RdY1Gn",title="Unemployement Rate in India")
figure.show()
```
